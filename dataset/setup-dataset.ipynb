```python
#Importing the pandas library for reading
import pandas as pd
```

```python
# load the CSV file into a Pandas dataframe
data = pd.read_csv("train.csv")

# print the first few rows of the dataframe
data.head()
```

```python
#Binarize identities columns

#Tags

#Gender
gender_columns = ['male','female','transgender', 'other_gender','heterosexual','homosexual_gay_or_lesbian', 'bisexual', 'other_sexual_orientation']
data[gender_columns] = data[gender_columns].astype(int)

#Religion
religion_columns = ['christian', 'jewish', 'muslim', 'hindu', 'buddhist', 'atheist', 'other_religion']
data[religion_columns] = data[religion_columns].astype(int)

#Race
race_columns = ['black', 'white', 'asian', 'latino', 'other_race_or_ethnicity']
data[race_columns] = data[race_columns].astype(int)

#Disability
disability_columns = ['physical_disability','intellectual_or_learning_disability','psychiatric_or_mental_illness', 'other_disability']
data[disability_columns] = data[disability_columns].astype(int)
```

```python
#Binarize bias subtypes 

biasTypes=['severe_toxicity','identity_attack','insult','obscene','sexual_explicit','threat']

for biasType in biasTypes:
    data[biasType]=data[biasType].apply(lambda x: 0 if x < 0.5 else 1)

def setOtherColumn(data, columns):
    value = 0
    for c in columns:
        value = value or data[c]
        if value == 1: break
    return 1 if value==0 else 0 

data['other'] = data.apply(setOtherColumn, args=(biasTypes,),axis=1)
```

```python
#merge certain columns into gender, religion, race, disability
def mergeColumns(data, columns):
    value = 0
    for c in columns:
        value = value or data[c]
        if value == 1: break
    return value
        
data['gender'] = data.apply(mergeColumns, args=(gender_columns,),axis=1)
data['religion'] = data.apply(mergeColumns, args=(religion_columns,),axis=1)
data['race'] = data.apply(mergeColumns, args=(race_columns,),axis=1)
data['disability'] = data.apply(mergeColumns, args=(disability_columns,),axis=1)
```

```python
data['Target']=data['target'].apply(lambda x: 0 if x < 0.4 else 1)
```

```python
#drop merged columns
data.drop(gender_columns, axis=1, inplace=True)
data.drop(religion_columns, axis=1, inplace=True)
data.drop(race_columns, axis=1, inplace=True)
data.drop(disability_columns, axis=1, inplace=True)
#data.drop('toxicity', axis=1, inplace=True)
```

```python
# Calculate the percentage of 0s and 1s in the "Target" column
target_counts = data['Target'].value_counts()
target_percentages = target_counts / target_counts.sum() * 100

# Print the percentage of 0s and 1s in the "Target" column
print("Percentage of 0s: {:.2f}%".format(target_percentages[0]))
print("Percentage of 1s: {:.2f}%".format(target_percentages[1]))
```
