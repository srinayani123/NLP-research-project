```python
import pickle
import pandas as pd
from sklearn.metrics import classification_report, accuracy_score
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Load the pickled model
with open('/kaggle/input/non-transformer-pickels/bilstm_base.pkl', 'rb') as file:
    model = pickle.load(file)

# Load the test data
test_df = pd.read_csv('/kaggle/input/bias-dataset-custom/test.csv')

# Drop empty rows
test_df.dropna(inplace=True)
```

```python
# Extract the text data and labels
X_test = test_df['comment_text'].values
y_test = test_df[['severe_toxicity', 'obscene', 'sexual_explicit', 'identity_attack', 'insult', 'threat', 'gender', 'religion', 'race', 'disability', 'Target']].values

# Tokenize and pad the sequences
tokenizer = Tokenizer(num_words=10000, oov_token="<OOV>")
tokenizer.fit_on_texts(X_test)
X_test_seq = tokenizer.texts_to_sequences(X_test)
X_test_pad = pad_sequences(X_test_seq, maxlen=200, padding='post')
```

```python
# Evaluate the model on the test data
loss, accuracy = model.evaluate(X_test_pad, y_test, verbose=0)

# Print the loss and accuracy
print("Loss: ", loss)
print("Accuracy: ", accuracy)
```
